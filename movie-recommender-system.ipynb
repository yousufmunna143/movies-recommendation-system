```python
# genres column is looking very weird.Basically it is list of dicitionaries
# The iloc method in Pandas is used for integer-based indexing. This means that you can use it to select rows and columns from a DataFrame by their integer positions. The syntax for using iloc is:

movies.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# i need this as
# ["Action", "Adventure" , "Fantasy" , "Science Fiction"]
# To perform above task lets create a function called convert
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

# for converting cast, first 3 members are enough
def convert_cast(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
            counter+=1
    return L

# for crew only director is enough
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
# overview is a string. we are converting into list so that appending can be easy with other lists
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# a small transformation needs to be applied. That is spaces between words needs to be removed
# This is because if there are spaces between words , they will be considered as different entities
# Eg:
# Sam Worthington(cast of avatar movie) and Sam Mendes(director of spectre movie) 
# These two are different persons and if model is created without removing spaces then 
# model gets confused whenever it comes across sam word
```

```python
movies['genres'] = movies['genres'].apply( lambda x:[ i.replace(" ","") for i in x ] )
movies['keywords'] = movies['keywords'].apply( lambda x:[ i.replace(" ","") for i in x ] )
movies['cast'] = movies['cast'].apply( lambda x:[ i.replace(" ","") for i in x ] )
movies['crew'] = movies['crew'].apply( lambda x:[ i.replace(" ","") for i in x ] )
```

```python
movies.head()
```

```python
# lets create a new column tags which is concatenation of overview, keywords, cast and crew columns

movies['tags'] = movies['overview']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
# lets create a new data frame with only movie_id, title and tags columns
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head()
```

```python
# converting tags columns from list to string
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
# converting everything in tags to lowercase letters
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
# stemmer is used to club similar words and replace with root words
# for eg., loved,loving,love are treated as love
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
stem(new_df['tags'][0])
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
# now we need to convert text data into features.
# First we will select most used 5000 words (features) and check how many times they are being repeated for each movie
# we will be using CountVectorizer class from sklearn module
# stop words(words which are used to form a sentence eg:is,and,or etc.) will not be considered as features
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
# checking features created
cv.get_feature_names_out()
```

```python
# now we have 5000 vectors
# task is to find distance between vectors.
# there are basically two types of distanceS, one is euclidian distance(sqrt((x1-x2)^2)+(x1-x2)^2)) and other is cosine distance(angle between vectors)
# but as number of dimensions increases finding euclidian distance becomes difficult and inaccurate
# hence, we will calculate cosine distances between vectors
# the higher the distances between vectors the less the similarity between them
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```
